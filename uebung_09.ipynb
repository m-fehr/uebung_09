#  Umrechnung zwischen Binär- und Dezimalsystem


<details style="border: 1px solid #bce8f1; background-color: #d9edf7; padding: 10px 15px; border-radius: 4px; margin: 1em 0;">
  <summary style="font-weight: bold; color: #31708f; cursor: pointer;">Umrechnung von Binär → Dezimal</summary>
  <p style="color: #333333;">
Wir wollen die Binärzahl  

$$
\texttt{10101}_2
$$

in das Dezimalsystem umwandeln.

**Rechenweg:**  
Jede Stelle der Binärzahl steht für eine Zweierpotenz (von rechts nach links):

$$
\begin{array}{cccccc}
\text{Stelle:} & 1 & 0 & 1 & 0 & 1 \\
\text{Potenz:} & 2^4 & 2^3 & 2^2 & 2^1 & 2^0 \\
\end{array}
$$

Nun multiplizieren wir jede Stelle mit der entsprechenden Potenz von 2:

$$
\begin{aligned}
1 \cdot 2^4 &= 16 \\
0 \cdot 2^3 &= 0 \\
1 \cdot 2^2 &= 4 \\
0 \cdot 2^1 &= 0 \\
1 \cdot 2^0 &= 1
\end{aligned}
$$


Dann addieren wir die Ergebnisse:

$$
16 + 0 + 4 + 0 + 1 = \boxed{21}
$$

**Ergebnis:**  
$$
\texttt{10101}_2 = \boxed{21_{10}}
$$
  </p>
</details>

<details style="border: 1px solid #bce8f1; background-color: #d9edf7; padding: 10px 15px; border-radius: 4px; margin: 1em 0;">
  <summary style="font-weight: bold; color: #31708f; cursor: pointer;">Umrechnung von Dezimal → Binär</summary>
  <p style="color: #333333;">

**Beispiel:**  
Wir wollen die Dezimalzahl

$$
\texttt{43}_{10}
$$  

ins Binärsystem umwandeln.

**Rechenweg (durch wiederholtes Teilen durch 2):**

$$
\begin{array}{rl}
43 \div 2 &= 21 \quad \text{Rest: } 1 \\
21 \div 2 &= 10 \quad \text{Rest: } 1 \\
10 \div 2 &= 5  \quad \text{Rest: } 0 \\
5 \div 2  &= 2  \quad \text{Rest: } 1 \\
2 \div 2  &= 1  \quad \text{Rest: } 0 \\
1 \div 2  &= 0  \quad \text{Rest: } 1 \\
\end{array}
$$

Nun lesen wir die Reste **von unten nach oben**:

$$
\boxed{101011}_2
$$

**Ergebnis:**  
$$
\texttt{43}_{10} = \boxed{101011_2}
$$
  </p>
</details>

<details style="border: 1px solid #bce8f1; background-color: #d9edf7; padding: 10px 15px; border-radius: 4px; margin: 1em 0;">
  <summary style="font-weight: bold; color: #31708f; cursor: pointer;">Umwandlung einer Dezimalzahl in ein 8-Bit-Gleitkommaformat (Mini-Float)
</summary>
  <p style="color: #333333;">
Wir wandeln die Dezimalzahl $0{,}75$ in eine 8-Bit-Gleitkommazahl um. Dieses vereinfachte Format (Mini-Float) wird oft für Lernzwecke verwendet.

---

## Mini-Float-Struktur (8 Bit)

| Feld       | Länge | Beschreibung                                  |
|------------|--------|-----------------------------------------------|
| Vorzeichen | 1 Bit  | 0 = positiv, 1 = negativ                      |
| Exponent   | 3 Bit  | mit **Bias = 3** (also tatsächlicher Wert = Code − 3) |
| Mantisse   | 4 Bit  | mit implizitem `1.` davor (Normierung)       |

---

## Schritt 1: Umwandlung der Dezimalzahl in Binär

### Nachkommateil: $0{,}75_{10}$

Wir wandeln den Nachkommateil durch **Multiplikation mit 2** um:

| Schritt | Nachkommazahl · 2      | Ganzzahl (Bit) | Neuer Nachkommateil |
|--------:|------------------------|----------------|----------------------|
| 1       | $0{,}75 \cdot 2 = 1{,}5$ | **1**          | $0{,}5$               |
| 2       | $0{,}5 \cdot 2 = 1{,}0$  | **1**          | $0{,}0$               |

Daraus ergibt sich:

$$
0{,}75_{10} = 0{,}11_2
$$

---

## Schritt 2: Normalisierung

Wir bringen die Binärzahl in die Form:

$$
1{,}1 \cdot 2^{-1}
$$

---

## Schritt 3: Bitfelder belegen

| Komponente      | Wert                          | Binär       |
|-----------------|-------------------------------|-------------|
| Vorzeichen      | Positiv → $0$                 | `0`         |
| Exponent        | $-1 + 3 = 2$                  | `010`       |
| Mantisse (nach dem Komma) | `1` → $1000$ (4 Bit, mit Nullen aufgefüllt) | `1000` |

---

## Ergebnis: 8-Bit-Mini-Float

$$
\texttt{0\ 010\ 1000}
$$

| Bitposition   | Bedeutung    |
|---------------|--------------|
| `0`           | Vorzeichen   |
| `010`         | Exponent     |
| `1000`        | Mantisse     |

---

## Rückrechnung zur Kontrolle

- Exponent = `010` = $2$ → tatsächlicher Exponent: $2 - 3 = -1$
- Mantisse: $1{,}1000_2 = 1{,}5$
- Wert:
  $$
  1{,}5 \cdot 2^{-1} = \boxed{0{,}75}
  $$

Die Rückrechnung bestätigt: Das Ergebnis stimmt.

---

## Merke

- **Nachkommazahlen** werden per Multiplikation mit 2 umgewandelt
- Das **erste Bit vor dem Komma wird nicht gespeichert** – es ist implizit (normiert)
- Der **Exponent** wird über einen **Bias** codiert (hier: Bias = 3)
- Die Mantisse wird auf eine feste Anzahl Bits **abgeschnitten oder gerundet**
  </p>
</details>

<details style="border: 1px solid #bce8f1; background-color: #d9edf7; padding: 10px 15px; border-radius: 4px; margin: 1em 0;">
  <summary style="font-weight: bold; color: #31708f; cursor: pointer;">Umwandlung einer 8-Bit-Gleitkommaformat-Zahl in eine Dezimalzahl</summary>
  <p style="color: #333333;">

| Feld        | Bits    | Bedeutung                     |
|-------------|---------|-------------------------------|
| Vorzeichen s| `1`     | negativ                       |
| Exponent e  | `101`   | binär `5` → $5 - 3 = 2$       |
| Mantisse m  | `0100`  | ergibt $1{,}0100_2 = 1{,}25$   |

---

## Schrittweise Rückrechnung

1. **Vorzeichenbit:**  
   `1` → Zahl ist **negativ**

2. **Exponent:**  
   $101_2 = 5$, Bias = 3 → tatsächlicher Exponent:
   $$
   5 - 3 = 2
   $$

3. **Mantisse:**  
   Durch die implizite führende `1` ergibt sich:
   $$
   1{,}0100_2 = 1 \cdot 2^0 + 0 \cdot 2^{-1} + 1 \cdot 2^{-2} + 0 \cdot 2^{-3} + 0 \cdot 2^{-4} = 1 + 0 + 0{,}25 = 1{,}25
   $$

4. **Gesamter Wert:** 
   $$ \mathsf z = (-1)^s \cdot m \cdot 2^e .$$

   $$
   \text{Wert} = -1{,}25 \cdot 2^2 = -1{,}25 \cdot 4 = \boxed{-5{,}0}
   $$

---

## Ergebnis:

Der 8-Bit-Wert `1 101 0100` entspricht der Dezimalzahl:

$$
\boxed{-5{,}0}
$$

---

## Übersicht der Umrechnung

| Bitfeld       | Inhalt     | Bedeutung                   |
|---------------|------------|-----------------------------|
| Vorzeichen    | `1`        | negativ                     |
| Exponent      | `101`      | $2$                         |
| Mantisse      | `0100`     | $1{,}25$                    |
| Wert berechnet| —          | $-1{,}25 \cdot 2^2 = -5{,}0$ |

  </p>
</details>
